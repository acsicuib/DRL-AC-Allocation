# ALL CASES

```python
fig, axs = plt.subplots(4, 2)

for axi,jobs in enumerate([9,12]):
    for axj,devices in enumerate([499,999,1499,1999]):
        CASE = devices+1
        print("CASE", CASE)
        data = pd.read_pickle('results/log_ga_mono_convergenceE'  + str(CASE) + "v"+str(jobs)+"_" + str(jobs) + '_' + str(devices)+'_0_w5050.pkl', 'infer')
        values = np.array(list(data))
        print(values.shape)
        axs[axj,axi].plot(values[:,0],)
        axs[axj,axi].title.set_text("Code: %i J: %i"%(CASE,jobs))
        

plt.show()
```
